# Figure 3: Threshold comparison of the mean field equilibrium with the mean field optimum for different values of c_V

```python
import matplotlib.pyplot as plt
import numpy as np
import math
import matplotlib.pyplot as plt
from mfg_computation import Parameters, mf_equilibrium, mf_optimum
```

```python
PARAMS = Parameters() # default parameters from paper
gamma = PARAMS.gamma # infection rate
rho = PARAMS.rho     # recovery rate
vac_min, theta = PARAMS.vac_min, PARAMS.theta # min/max vaccination rate
T = PARAMS.T # time horizon

c_V = PARAMS.c_V       # cost per unit time of vaccination
c_I = PARAMS.c_I      # cost per unit time of infection

# initial conditions
S0 = 0.4        # proportion of susceptible at t=0
I0 = 0.4        # proportion of infected at t=0
```

```python
v_cV = np.arange(0.1,1.1,0.05)

thr_mfe = []
thr_mfopt = []
    
for val in range(len(v_cV)):#v_cV:
    PARAMS.c_V = v_cV[val]
    cost_opt,t1 = mf_optimum(PARAMS, S0, I0, C=10000)
    cost_eq, t2 = mf_equilibrium(PARAMS, S0, I0, C=10000)
    thr_mfopt.append(t1)
    thr_mfe.append(t2)
    
plt.plot(v_cV,thr_mfopt,'k',label='OPT')
plt.plot(v_cV,thr_mfe,'--k',label='MFE')
plt.legend()#(fancybox=True, framealpha=1, shadow=True, borderpad=1)
plt.xlabel('Vaccination cost')
plt.ylabel('Jump time')
```
